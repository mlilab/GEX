```python
import jax
import jax.numpy as jnp
import haiku as hk
from functools import partial
from vit import ViT
```

```python
net = partial(ViT,
        patch_size = 32,
        num_classes = 1000,
        hidden_size = 384,
        mlp_dim = 1536,
        num_layers = 12,
        num_heads = 6,
        dropout_rate = 0.1,
        name='vit_s_32',
    )

def f(*args, **kwargs):
        return net()(*args, **kwargs)

f = hk.transform_with_state(f)

rng = jax.random.PRNGKey(42)
x = jnp.zeros([5, 224, 224, 3])
params, state = f.init(rng, x, train=True, print_shape=True)
for k,v in params.items():
    for k_,v_ in v.items():
        print(f'{k}_{k_}:{v_.shape}')
```
